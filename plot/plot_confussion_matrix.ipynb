```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind
import os
import sys
from tabulate import tabulate
import seaborn as sns
from matplotlib.colors import LogNorm

colors = ['#1F77B4', '#2CA02C', '#FF7F0E', '#D62728', '#535154', 'black', '#F6C85F']

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
datasets = ["Exp-IN/nops/", "Exp-OUT1/nops/", "Exp-OUT1/random/", "Exp-OUT2/nops/"]
```

```python
for dataset in datasets:

    confusion_matrix = pd.read_csv("/media/SSD/paper_data/sakura/"+dataset+"results/resnet_pnone_10000_s0-sensors-all_10fold/out/cm/LSTM+CNN_kf5confusion_metrix.csv", index_col=0)
    confusion_matrix = confusion_matrix.round(decimals=2)*100
    confusion_matrix[confusion_matrix == 0] = np.nan 
    
    #, norm=LogNorm()

    plt.figure(figsize = (20,20), dpi=80)
    ax = sns.heatmap(confusion_matrix, annot=True, center=0, cbar=False, cmap=sns.diverging_palette(220, 15, s=50, as_cmap=True), fmt='g', linewidths=0.5, linestyle='--', linecolor='lightgray', annot_kws={"fontsize":13, "fontweight":'bold'})#cmap='')
    ax.xaxis.set_ticks_position('top');

    ax.set_yticklabels(ax.get_ymajorticklabels(), fontsize = 22);
    ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 22);
    plt.yticks(rotation=0, weight='bold');
    plt.xticks(rotation=90, weight='bold');
    dataset_stripped = dataset.replace("/nops/", "-N").replace("/random/", "-R")
    plt.savefig('/media/SSD/paper_data/plots/fig-confusion-matrix-'+dataset_stripped+'-resnet.svg', dpi=300, bbox_inches = 'tight', pad_inches = 0)
```

```python
dataset = "Exp-OUT1/random/"
model = "resnet"


confusion_matrix = pd.read_csv("/media/SSD/paper_data//sakura/"+dataset+"/results/resnet_htop_pnone_10000_s0-sensors-all_10fold/out/cm/LSTM+CNN_kf5confusion_metrix.csv", index_col=0)
confusion_matrix = confusion_matrix.round(decimals=2)*100
confusion_matrix[confusion_matrix == 0] = np.nan
types = ['arith', 'logic', 'compare', 'shift', 'load', 'store', 'branch', 'jump']
confusion_matrix = confusion_matrix.loc[types, types]

#, norm=LogNorm()

plt.figure(figsize = (7,7), dpi=80)
ax = sns.heatmap(confusion_matrix, annot=True, center=0, cbar=False, cmap=sns.diverging_palette(220, 15, s=50, as_cmap=True), fmt='g', linewidths=0.5, linestyle='--', linecolor='lightgray', annot_kws={"fontsize":20, "fontweight":'bold'})#cmap='')
ax.xaxis.set_ticks_position('top');

ax.set_yticklabels(ax.get_ymajorticklabels(), fontsize = 22);
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 22);
plt.yticks(rotation=0, weight='bold');
plt.xticks(rotation=90, weight='bold');
dataset_stripped = dataset.replace("/nops/", "-N").replace("/random/", "-R")
plt.savefig('/media/SSD/paper_data/plots/fig-confusion-matrix-'+dataset_stripped+'-resnet-hierarchical.svg', dpi=300, bbox_inches = 'tight', pad_inches = 0)
plt.savefig('/media/SSD/paper_data/plots/fig-confusion-matrix-'+dataset_stripped+'-resnet-hierarchical.pdf', dpi=300, bbox_inches = 'tight', pad_inches = 0)
```
